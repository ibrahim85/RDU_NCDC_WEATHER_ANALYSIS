### Pull Weather Data from API 
#### Raleigh, Charlotte, DC 1946 - 2015

```python
import json
import requests
import pydotplus
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.svm import SVC
from matplotlib import pyplot as plt
from IPython.display import Image, display
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
%matplotlib inline

#Sites we will loop through
site_name=['RDU','Reagan_Intl', 'CLT']
site_id=['GHCND:USW00013722','GHCND:USW00013743', 'GHCND:USW00013881']

#Variable Shortcuts
inputs=['TAVG', 'CLDD', 'DX32', 'TMIN', 'TMAX', 'HTDD', 'DX90', 
        'DP01', 'DP05', 'DT32', 'FZF0', 'PRCP', 'RDU', 'Reagan_Intl']
labels=['AVG Temp C', 'Cooling Degree Days','# Days max temp <32F','MinTemp',
        'MaxTemp','Heating Degree Days','# Days max>90F','# Day > .1 inch Precip',
        '# Day > .5 inch of precip','# Day min tem < 32F','Temp first freeze < 32F',
        'Precpitation for year','RDU', 'Reagan_Intl']
targets=['Snow', 'No Snow']
```

### Pull data for a specific location for 10 year interval (max allowed by API in one call)

```python
def return_results(url):
    
    #Create dictionaries for variable
    columns=['Year','SNOW','TAVG','CLDD','DX32','TMIN','TMAX',
             'HTDD','DX90','DP01', 'DP05','DT32','FZF0','PRCP']
    all_col={}
    for column in columns:
        all_col[column] = []
   
    #Call API
    #headers = {'token': 'dKuJxEKlhIJuoZSjvKULivIPXWsRqspt' } #Two api keys, 1000 request limit/day
    headers = {'token': 'HGQwYMutaGbKyJUqOZchjtieVMKpMMTE'}
    response = requests.get(url, headers=headers)
    parsed=json.loads(response.text)   
       
    for record in parsed['results']:
        for i in range(1,len(columns)):
            if record['datatype'] == columns[i]: 
                if i==1:
                    all_col['Year'].append(record['date'])
                all_col[columns[i]].append(record['value'])
               
    return all_col             
```

### Pull data in 10 year intervals for a specific site and merge results

```python
def year_loop(start,end, by, code):
    
    columns=['Year','SNOW','TAVG','CLDD','DX32','TMIN','TMAX',
             'HTDD','DX90','DP01', 'DP05','DT32','FZF0','PRCP']
    all_col2={}
    for column in columns:
        all_col2[column] = []
        
    url='http://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GSOY&stationid=id_code&startdate=1946-01-01&enddate=1955-01-01&limit=1000'
    url=url.replace('id_code', code)

    ranges=np.arange(start, end+1, by)
    for i in range(len(ranges)-1):
        
        all_cols = return_results(url)

        for key, value in all_cols.items():
            all_col2[key].extend(value)
            
        if i==6:
            pass
            
        else:
            url=url.replace(str(ranges[i+1]-1),str(ranges[i+2]-1))
            url=url.replace(str(ranges[i]),str(ranges[i+1]))
        
    all_col2=pd.DataFrame.from_dict(all_col2)
    all_col2['Year']=all_col2['Year'].map(lambda x: x[:4])

    return all_col2
```

### Pull data for Raleigh, Charlotte and DC from 1946 - 2010 - Load to CAS 

```python
snow=pd.DataFrame()
for i in range(len(site_name)):
    temp=year_loop(1946,2016,10,site_id[i])
    temp['Location']=site_name[i]
    snow=pd.concat([snow, temp], axis=0)
```

```python
#Create snow variable
snow['NO_SNOW_FLAG'] = (snow['SNOW']<100).astype(int)
snow.head()
```

### Data Dictionary

```python
data_vars = inputs + ['NO_SNOW_FLAG', 'Year']
data_labels = labels + ['Less than 100mm Snow in Year', 'YEAR']

Data_dictionary = pd.DataFrame({
        'Variable': data_vars,
        'Description': data_labels
    })

Data_dictionary
```

### Encode Location Variable

```python
snow=pd.concat([snow, pd.get_dummies(snow['Location'],drop_first=False)], axis=1)
snow.head()
```

### Stratify Data

```python
stratified=StratifiedShuffleSplit(n_splits=1, test_size=.3, random_state=1234)
stratification=list(stratified.split(snow, snow['NO_SNOW_FLAG']))
snow['SPLIT']=(snow.index.isin(stratification[0][0])).astype(int)

X_train=snow[inputs][snow['SPLIT']==1]
Y_train=snow['NO_SNOW_FLAG'][snow['SPLIT']==1]
X_valid=snow[inputs][snow['SPLIT']==0]
Y_valid=snow['NO_SNOW_FLAG'][snow['SPLIT']==0]
```

## Build Models

```python
#%% capture
#decision tree
dt = DecisionTreeClassifier(max_depth=4, min_samples_split=5, criterion='gini', min_samples_leaf=2)
dt.fit(X_train, Y_train)
valid_dt_pred=dt.predict(X_valid)

#Random Forest
rf = RandomForestClassifier(n_estimators=100, oob_score=True, max_depth=5)
rf.fit(X_train, Y_train)
valid_rf_pred=rf.predict(X_valid)

#Gradient Boost
gb = GradientBoostingClassifier()
gb.fit(X_train, Y_train)
valid_gb_pred=gb.predict(X_valid)

#Logistic Regression
lm = LogisticRegression(C=1)
lm.fit(X_train, Y_train)
valid_lm_pred=lm.predict(X_valid)
```

### Score Models and get Variable Importance

```python
#Decision tree
dt_accuracy = accuracy_score(Y_valid, valid_dt_pred)
dt_importance = dt.feature_importances_

#Random Forest
rf_accuracy = accuracy_score(Y_valid, valid_rf_pred)
rf_importance = rf.feature_importances_

#Gradient Boost
gb_accuracy = accuracy_score(Y_valid, valid_gb_pred)
gb_importance = gb.feature_importances_

#Logistic Regression
lm_accuracy = accuracy_score(Y_valid, valid_lm_pred)

models = pd.DataFrame({ 'Name': ['Decision Tree','Random Forest',
                                 'Gradient Boost', 'Logistic Regression'],
                        'Misclassification Rate': [dt_accuracy, rf_accuracy, 
                                                  gb_accuracy, lm_accuracy]})
                                     
models['Misclassification Rate']= models['Misclassification Rate'].map(lambda x: '%.8f' %x)

models[['Name','Misclassification Rate']]
```

```python
var_import=pd.DataFrame({'Variables': inputs,
                         'Decision Tree': dt_importance,
                         'Random Forest': rf_importance,
                         'Gradient Boost': gb_importance})

var_import.sort_values(by='Random Forest', ascending=False)
```

```python
predictions=[valid_dt_pred, valid_rf_pred, valid_gb_pred, valid_lm_pred]
model_name=['dt','gb','rf','lm']
roc_data = {'fpr': [],
            'tpr': [],
            'Model': []}
roc=[]
for i, prediction in enumerate(predictions):
    temp = roc_auc_score(Y_valid, prediction)
    fpr, tpr, threshold = roc_curve(Y_valid, prediction)
    roc.append(temp)
    roc_data['fpr'].extend(fpr)
    roc_data['tpr'].extend(tpr)
    roc_data['Model'].extend([models['Name'][i] for y in range(len(fpr))])

roc_data=pd.DataFrame(roc_data)
roc_data
```

```python
plt.figure()
for key, grp in roc_data.groupby(["Model"]):
    plt.plot(grp["fpr"], grp["tpr"], label=key)
plt.plot([0,1], [0,1], "k--")
plt.xlabel("False Postivie Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="best")
plt.title("ROC Curve (using validation data)")
plt.show()
```
